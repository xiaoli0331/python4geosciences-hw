# Q2: Test for vowels

```python
def isvowels(a):
    if a in ['a','e','i','o','u','A','E','I','O','U']:
        return True
    else:
        return False
print(isvowels("A"))
print(isvowels("e"))
print(isvowels("F"))
print(isvowels("3"))
```

# Q3: Compute the length of a given list or string.

```python
def length(a):
    count = 0
    for b in a:
        count += 1
    return count
print(length('Xiao Li'))
print(length(['I','am','Xiao','Li']))
```

# Q5: Check for palindromes

```python
def is_palindrome(a):
    #remove all the spaces
    a=a.replace(" ","") 
    #convert the string to lowercase & check for palindromes
    return a.lower() == a.lower()[::-1] 
print(is_palindrome("Never odd or even"))
print(is_palindrome("This is Xiao"))
```

# Q.8 Encode and decode using ROT-13

```python
key = {'a':'n', 'b':'o', 'c':'p', 'd':'q', 'e':'r', 'f':'s', 'g':'t', 'h':'u',
       'i':'v', 'j':'w', 'k':'x', 'l':'y', 'm':'z', 'n':'a', 'o':'b', 'p':'c', 
       'q':'d', 'r':'e', 's':'f', 't':'g', 'u':'h', 'v':'i', 'w':'j', 'x':'k',
       'y':'l', 'z':'m', 'A':'N', 'B':'O', 'C':'P', 'D':'Q', 'E':'R', 'F':'S', 
       'G':'T', 'H':'U', 'I':'V', 'J':'W', 'K':'X', 'L':'Y', 'M':'Z', 'N':'A', 
       'O':'B', 'P':'C', 'Q':'D', 'R':'E', 'S':'F', 'T':'G', 'U':'H', 'V':'I', 
       'W':'J', 'X':'K', 'Y':'L', 'Z':'M'}
def En_De(a):
    sentence = ''
    for b in a:
        if b in key.keys():
            sentence += key[b]
        else:
            sentence += b
    return sentence

print(En_De("Pnrfne pvcure? V zhpu cersre Pnrfne fnynq!"))
print(En_De('Caesar cipher? I much prefer Caesar salad!'))
```
