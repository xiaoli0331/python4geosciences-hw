```python
from datetime import datetime
f = open('/Users/lixiao/python4geosciences/data/GPS.dat')
f.seek(0)
drifter= ''
Date=[]
a=f.readlines()
for line in a:
    data = line.split('\t')
    if not data: continue
    if data[0] == 'Track':
        drifter = data[1]
        continue
    if drifter =='ACTIVE LOG':
        if data[0] == 'Trackpoint':
            Datetime = data[2].split()
            Year = int(Datetime[0].split('/')[2])
            Month = int(Datetime[0].split('/')[0])
            Day = int(Datetime[0].split('/')[1])
            Hour = int(Datetime[1].split(':')[0])
            Minute = int(Datetime[1].split(':')[1])
            Second = int(Datetime[1].split(':')[2])
            if Datetime[2] == 'PM':
                Hour = Hour + 12
            Date.append(datetime(Year,Month,Day,Hour,Minute,Second))
Date
```
